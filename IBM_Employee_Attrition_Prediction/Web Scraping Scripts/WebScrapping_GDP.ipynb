**Web Scrapping for database creation of Economy of Location**

Import required libraries

```
import urllib3
from bs4 import BeautifulSoup
```

Specify the URL on which web scrapping is to be performed.

```
url = 'https://en.wikipedia.org/wiki/List_of_California_locations_by_income#Entire_state'
```

Query the website and return the HTML to the variable webpage.

```
http = urllib3.PoolManager()
webpage = http.request('GET', 'https://en.wikipedia.org/wiki/List_of_California_locations_by_income#Entire_state')
soup = BeautifulSoup(webpage.data)
```

Find all tables from Beatutiful soup

```
all_tables=soup.find_all("table")
all_tables
```

Select the required table.

```
req_table=soup.find('table', class_='wikitable sortable')
req_table
```

Store the cells from all table columns in different lists.

```
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]

for row in req_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==6:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
        D.append(cells[3].find(text=True))
        E.append(cells[4].find(text=True))
        F.append(cells[5].find(text=True))
```

Storing the data from python lists in Pandas Dataframe.

```
import pandas as pd
df=pd.DataFrame(A,columns=['County'])
df['Population']=B
df['Population Density']=C
df['Per Capita Income']=D
df['Median Household Income']=E
df['Median Family Income']=F
df
```

We will perform data cleaning in next step. Here we will remove the special symbols from the columns.

```
#Removing special symbols from dataset columns
special_symbols = ['$', ',', '\n']

for i in range(0,len(data['Population'])):
  data['Population'][i] = data['Population'][i].replace(',','')

for i in range(0,len(data['Per Capita Income'])):
  for n in special_symbols:
    data['Per Capita Income'][i] = data['Per Capita Income'][i].replace(n,'')

for i in range(0,len(data['Median Household Income'])):
  for n in special_symbols:
    data['Median Household Income'][i] = data['Median Household Income'][i].replace(n,'')

for i in range(0,len(data['Median Family Income'])):
  for n in special_symbols:
    data['Median Family Income'][i] = data['Median Family Income'][i].replace(n,'')
```

In next step, we will perform feature engineering to generate new feature 'GDP' in our GDP dataset. We will use below formula: 

*GDP = Population * Per Capita Income*

To perform feature engineering, we will change the datatype of features Population and Per Capita Income.


```
#Changing datatype of features
for i in range(0,len(data['Population'])):
  data['Population'][i] = int(data['Population'][i])

for i in range(0,len(data['Per Capita Income'])):
  for n in special_symbols:
    data['Per Capita Income'][i] = int(data['Per Capita Income'][i])
```

In next step, we will perform feature engineering to generate new feature 'GDP'.

```
#Feature Generation from existing columns of dataset
data['GDP'] = data['Population'] * data['Per Capita Income']
```
